Who Wants to Be a Millionaire is a popular game show that originated in the United Kingdom in 1998 and has been adapted in over 160 countries worldwide. The objective of the game is for contestants to correctly answer a series of increasingly difficult multiple-choice questions in order to win a cash prize, with the ultimate goal of becoming a "millionaire" by correctly answering the final question.

The show is typically hosted by a celebrity and features a contestant, who is given the opportunity to use "lifelines" such as "50/50" (eliminating two incorrect answers), "Phone a Friend" (calling a pre-approved friend for help), and "Ask the Audience" (polling the live studio audience) in order to assist them in answering the questions.

The format of the show is designed to be high-stakes and dramatic, with a ticking clock and increasing levels of difficulty as the contestant progresses through the questions. The show has been very successful and has been credited with popularizing the concept of the "fastest finger" question, in which contestants must answer a question correctly and quickly in order to advance to the next round.

As reported by the New York Times, the show's creator and executive producer, David Briggs, had the idea of "Who wants to be a millionaire" while he was working on a documentary about the United Kingdom's National Lottery. He thought of using the format of a trivia game show with a top prize of a million pounds to create a similar sensation of excitement and anticipation.
